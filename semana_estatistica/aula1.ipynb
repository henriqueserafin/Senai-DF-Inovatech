- loc() -  Pesquisa pelo label

````python
# extrair a série de dados do Brasil
brasil = df.loc['Brasil',anos]
````

````python
brasil
````

````python
type(brasil)
````

````python
# tolist
brasil_dict={'ano':brasil.index.tolist(),'imigrantes':brasil.values.tolist()}
type(brasil_dict)
````

````python
df_brasil = pd.DataFrame(brasil_dict)
````

# Criando o gráfico de linhas

````python
plt.plot(df_brasil['ano'],df_brasil['imigrantes'],)
plt.xticks(['1980','1985','1990','1995','2000','2005','2010'])
plt.yticks([500,1000,1500,2000,2500,3000])
plt.grid()
plt.show()
````

````python
plt.figure(figsize=(8,4))
plt.plot(df_brasil['ano'],df_brasil['imigrantes'])
plt.title("Imigração do Brasil para o Canadá")
plt.xlabel('Ano')
plt.ylabel('Número de imigrantes')
plt.xticks(['1980','1985','1990','1995','2000','2005','2010','2013'])
plt.show()
````

````python
fig , ax = plt.subplots(figsize=(8,4))
ax.plot(df_brasil['ano'],df_brasil['imigrantes'],lw=5,color='#FE5000',marker='o')
ax.set_title('Imigração do Brasil para o Canadá \n 1980 a 2013', loc='left')
ax.xaxis.set_major_locator(plt.MultipleLocator(5))
ax.set_xlabel('Ano')
ax.set_ylabel('Número de imigrantes')
plt.grid(linestyle='--')
plt.show()
````

# Gráfico de linha e boxplot em subplots

````python
fig,axs = plt.subplots(1,2,figsize=(15,5))
# subplot 1 - gráfico linha
axs[0].plot(df_brasil['ano'],df_brasil['imigrantes'])
axs[0].set_title('Imigração do Brasil para o Canadá \n 1980 a 2013')
axs[0].xaxis.set_major_locator(plt.MultipleLocator(5))
axs[0].set_xlabel('Ano')
axs[0].set_ylabel('Número de imigrantes')
axs[0].grid()
# subplot 2 - gráfico boxplot
axs[1].boxplot(df_brasil['imigrantes'])
axs[1].set_title("Box-plot da imigração do Brasil para o Canadá\n 1980 a 2013")
axs[1].set_xlabel('Brasíl')
axs[1].set_ylabel('Número de imigrantes')
axs[1].grid()
````

# Comparativo dos maiores paises da America Latina (imigração Canadá)

````python
# Criar uma figura
fig,axs = plt.subplots(2,2,figsize=(10,6))
# Brasil
axs[0,0].plot(df.loc['Brasil',anos])
axs[0,0].set_title('Brasil')

# Colombia
axs[0,1].plot(df.loc['Colômbia',anos])
axs[0,1].set_title('Colômbia')

# Argentina
axs[1,0].plot(df.loc['Argentina',anos])
axs[1,0].set_title('Argentina')

# Peru
axs[1,1].plot(df.loc['Peru',anos])
axs[1,1].set_title('Peru')

fig.subplots_adjust(hspace=0.5,wspace=0.6)

for ax in axs.flat:
  ax.xaxis.set_major_locator(plt.MultipleLocator(5))

for ax in axs.flat:
  ax.set_xlabel('Ano')
  ax.set_ylabel('Números de Imigrantes')


ymin = 0
ymax = 7000
# criar um for usar ravel() - achatar uma matriz bi para uma matriz uni
for ax in axs.ravel(): # uso do array unidminensional
  # usar set_ylim() - determinar o mínimo e o máximo do eixo y
  ax.set_ylim(ymin,ymax)
fig.suptitle("imigração na América Latina",fontsize=16)
````

````python
america_sul=df.query('Região =="América do Sul" ')
````

````python
america_sul
````

````python
cores = ['silver', 'orange', 'forestgreen', 'orchid', 'purple', 'brown', 'slateblue', 'gray', 'olive', 'navy', 'teal', 'tomato']
````

````python
america_sul_sorted=america_sul.sort_values(by='Total',ascending=True)

display(america_sul_sorted)
````

````python
cores=[]

for pais in america_sul_sorted.index:
    if pais == 'Brasil':
        cores.append('green')
    else:
        cores.append('silver')

fig,ax=plt.subplots(figsize=(12,5))
ax.barh(america_sul_sorted.index,america_sul_sorted['Total'],color=cores)
ax.set_title('Imigração da América do Sul para o Canadá')
ax.set_xlabel('Número de Imigrantes',fontsize=14)
ax.set_ylabel('')
ax.xaxis.set_tick_params(labelsize=12)
ax.yaxis.set_tick_params(labelsize=12)
plt.show()
````

````python
for i,v in enumerate(america_sul_sorted['Total']):
    print(f'o valor de i: {i} v: {v}')
````

````python
cores=[]

for pais in america_sul_sorted.index:
    if pais == 'Brasil':
        cores.append('green')
    else:
        cores.append('silver')

fig,ax=plt.subplots(figsize=(12,5))
ax.barh(america_sul_sorted.index,america_sul_sorted['Total'],color=cores)
ax.set_title('Imigração da América do Sul para o Canadá 1980 a 2013')
ax.set_xlabel('Número de Imigrantes',fontsize=14)
ax.set_ylabel('')
ax.xaxis.set_tick_params(labelsize=12)
ax.yaxis.set_tick_params(labelsize=12)
for i,v in enumerate(america_sul_sorted['Total']):
    # print(f'o valor de i: {i} v: {v}')
    ax.text(v,i,str(v),fontsize=12)
ax.set_frame_on(False)
ax.get_xaxis().set_visible(False)
plt.show()
````
