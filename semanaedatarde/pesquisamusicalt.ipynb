#### tarefa 1: exportar a planilha como .csv

#### tarefa 2: importar o .csv no pandas

```python
df=pd.read_csv(url)
```

#### tarefa 3: remover a primeira coluna e renomear as outras

```python
df.columns = ['data','nota_rock','nota_hiphop','nota_mpb','nota_reggae','nota_funk','nota_sertanejo','nota_gospel','nota_eletronica']
df
```

```python
del df['data']
```

#### tarefa 4: escolher um gráfico que possa mostrar para a professora e o orquestrador de IA quais os estilos tiveram um nota maior ou menor.

```python
df.sum().plot(kind='bar', figsize=(10, 6), title="Total de Votos por Estilo Musical")
```
