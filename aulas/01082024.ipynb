### Usamos char e varchar na seguinte situação, se euy uso char(8) apenas valores com 8 caracteres vão entrar no banco de dados, mas caso eu uso varchar(8) pode entrar algo de 5 caracteres, já que pode entrar qualquer valor, limitado à 8 caractetres
utilizando o comando abaixo podemos mostrar o diario de bordo com o ra e id do aluno, já que este join mostra todas as colunas das duas tabelas
```sql
select * from diariobordo join aluno on diariobordo.fk_aluno_id=aluno.id;
```
