## Joins 😱

Para fazer um join em SQL é necessário usar a palavra-chave JOIN seguida do nome da tabela que você deseja unir e a palavra-chave ON seguida do nome da coluna que contém a chave primária da primeira tabela e a chave estrangeira da segunda tabela, respectivamente.

Abaixo um exemplo de um join simples entre a tabela diariobordo e a aluno:


```sql
select 
	d.id,
	d.texto,
	d.datahora,
	a.ra
    -- nesse caso utilizamos o select para pegar as colunas que vamos usar, a forma para escrever é o nome da tabela ponto nome da coluna, para facilitar vamos chamar diariobordo de d e aluno de a
from
	diariobordo as d
join
	aluno as a
on
	d.fk_aluno_id=a.id;
    -- aqui estamos dizendo que a chave estrangeira do diario é igual a chave primária, ou id, do aluno
```
