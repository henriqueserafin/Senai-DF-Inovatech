#### separamos o querry com DDL  DML DQL DCL DTL. que são 5 tipos de SQL A explicação é a seguinte:


### DDL (Data Definition Language)

DDL refere-se à linguagem de definição de dados. É usada para definir e modificar a estrutura dos objetos de banco de dados, como tabelas, índices e esquemas. Comandos DDL não manipulam dados em si, mas sim a estrutura do banco de dados. Os principais comandos DDL incluem:

- **CREATE**: Cria novos objetos no banco de dados, como tabelas, índices e esquemas

```sql
CREATE TABLE alunos (
id INT PRIMARY KEY,
nome VARCHAR(100),
idade INT
);
```

- **ALTER**: Modifica a estrutura de objetos existentes no banco de dados, como adicionar uma nova coluna a uma tabela.

```sql
    ALTER TABLE alunos ADD email VARCHAR(100);
```

- **DROP:** Remove objetos do banco de dados, como tabelas ou índices.

```sql
    DROP TABLE alunos;
```

- **TRUNCATE**: Remove todos os registros de uma tabela, mas mantém sua estrutura.

```sql
    TRUNCATE TABLE alunos;
```


### DML (Data Manipulation Language)

DML refere-se à linguagem de manipulação de dados. É usada para inserir, atualizar, deletar e manipular dados dentro das tabelas. Os principais comandos DML incluem:

- **INSERT**: Insere novos registros em uma tabela.
    
    ```sql
    INSERT INTO alunos (id, nome, idade) VALUES (1, 'João', 20);
    
    ```
    
- **UPDATE**: Atualiza dados existentes em uma tabela.
    
    ```sql
    UPDATE alunos SET idade = 21 WHERE id = 1;
    
    ```
    
- **DELETE**: Remove registros de uma tabela.
    
    ```sql
    DELETE FROM alunos WHERE id = 1;
    
    ```

### DQL (Data Query Language)

DQL refere-se à linguagem de consulta de dados. É usada para consultar e buscar informações nas tabelas. O principal comando DQL é:

- **SELECT**: Recupera dados de uma ou mais tabelas.
    
    ```sql
    SELECT nome, idade FROM alunos WHERE idade > 18;
    
    ```

### DCL (Data Control Language)

DCL refere-se à linguagem de controle de dados. É usada para controlar o acesso aos dados no banco de dados, definindo permissões e segurança. Os principais comandos DCL incluem:

- **GRANT**: Concede permissões a usuários ou roles para realizar determinadas operações no banco de dados.
    
    ```sql
    GRANT SELECT, INSERT ON alunos TO usuario;
    
    ```
    
- **REVOKE**: Revoga permissões previamente concedidas a usuários ou roles.
    
    ```sql
    REVOKE INSERT ON alunos FROM usuario;
    
    ```

### DTL (Data Transaction Language)

DTL refere-se à linguagem de transação de dados. É usada para gerenciar transações no banco de dados, garantindo a consistência e integridade dos dados. Os principais comandos DTL incluem:

- **BEGIN TRANSACTION**: Inicia uma nova transação.
    
    ```sql
    BEGIN TRANSACTION;
    
    ```
    
- **COMMIT**: Confirma as operações realizadas na transação, tornando-as permanentes no banco de dados.
    
    ```sql
    COMMIT;
    
    ```
    
- **ROLLBACK**: Reverte as operações realizadas na transação, retornando o banco de dados ao estado anterior ao início da transação.
    
    ```sql
    ROLLBACK;
    
    ```

### Esses conceitos e comandos são fundamentais para a administração e manipulação de dados em sistemas de bancos de dados relacionais, permitindo que os usuários definam, consultem, modifiquem, controlem o acesso e garantam a integridade dos dados.
